# Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/andres-patrignani/Spring-2020/master/Datasets/mauna_loa_co2.csv')
df.head()
```

```python
df.tail()
```

```python
# Note that datetime works only in the format of day, month, and year
df['decimal_date'] = df['year']+ df['month']/12

df['decimal_date']
```

```python
plt.figure(figsize = (12,8))
plt.plot(df['decimal_date'], df['co2_ppm'])
plt.show
```

```python
# Scatter plot

plt.scatter(df['decimal_date'], df['co2_ppm'])
```

```python
# looking for CO2 for dates greater than 2010

idx = df['decimal_date']> 2010
print(idx)
```

```python
plt.figure()
plt.plot(df['decimal_date'][idx], df['co2_ppm'][idx])
plt.ylabel('CO$_2$ concentration (ppm)', size = 16)
plt.xticks(fontsize = 16, fontweight = 'bold') # Resizing the axis tixk marks
plt.yticks(fontsize = 16, fontweight = 'bold')
```
