# Newton's Law of Cooling

```python
# dT = -K (Tobs - Tambient) * t

#Initial condition
T_cup_pred = [68] # degrees Celcius. We used [] because we want to designate 68 as a list 
# otherwise, we cannot use the append command as it does not work for integers.
Tamb = 25.4 # degrees Celcius
dt = 1 # minutes
period = 121 # minutes
t_cup_pred = range(period) #in minutes
K = 0.02
```

```python
# Implement the model
for t in range(1,period):
    dT = -K * (T_cup_pred[t-1] - Tamb) * dt
    T_cup_pred_updated = T_cup_pred[t-1] + dT
    T_cup_pred.append(T_cup_pred_updated)
    #print(T_cup_pred)
```

```python
t_cup_obs = [0,3,6,9,12,18,24,30,40]
T_cups_obs = [68,64,60.3,56.8,54.5,49.6,46.8,44.1,40.3]
```

```python
import matplotlib.pyplot as plt  # matplotlib is the most popular library for plotting graphs in python

# %matplotlib inline #This line must be added to told versions of anaconda for the plot to work

plt.figure(figsize=(8,5))
plt.plot(t_cup_pred, T_cup_pred,'--k')
plt.plot(t_cup_obs, T_cups_obs, 'ok')
plt.xlabel('Time (minutes)', size = 16)
plt.ylabel('Temperature (Celcius)',size = 16)
plt.show()
```
